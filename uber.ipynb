```python
#REMOVE OUTLIERS 
q_low=df["fare_amount"].quantile(0.01)
q_high=df["fare_amount"].quantile(0.99)
df=df[(df["fare_amount"]>q_low) & (df["fare_amount"]<q_high)]
```

```python
#Applying linear regression models
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop("fare_amount",axis=1)
y=df["fare_amount"]
```

```python
x["pickup_datetime"]=pd.to_numeric((pd.to_datetime(df["pickup_datetime"])))
x = x.loc[:, x.columns.str.contains('^Unnamed')]
```

```python
#DIVIDING THE DATA SET INTO TEST TRAIN SPLIT 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#Applying to linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#PREDICTION
predict=lr.predict(x_test)
```

```python
#rmse
from sklearn.metrics import root_mean_squared_error
rmse=root_mean_squared_error(predict,y_test)
print("RMSE:",rmse)
```
